## Introduction
During the Data Wrangling process where datasets were gathered, assessed, cleaned and merged to create `archive master` dataset.The dataset was then split into two sub-dataframes i.e `tweets_df` and `images_df` for analysis. The following topics were considered to gain insights from the data:

* Dynamic between Favorite Counts and Retweet Counts
* Accuracy of prediction being a valid dog breed
* Most popular source of tweets

## EDA and Visualization

#### 1. Dynamic between Favourite Counts and Retweet Counts
The descriptive statistics of `tweet_df` highlighted a right-skewed distribution for retweet and favourite counts. Hence, the need to use the counts of favourites and retweets to understand the popularity of the tweets.

The scatter plot showed that the retweet counts and favorite counts of Tweets have a positive correlation. Thus, a popular Tweet with a large number of favorite counts (or retweet counts) is expected to have a relatively large number of retweet counts (or favorite counts).


#### 2. Accuracy of Prediction being a Valid Dog Breed
The neural network's accuracy in correctly identifying images of dogs in the three prediction tries was analyzed. Observation of the bar plot confirmed the means of prediction levels given how closely similar the proportions of the blue and orange areas are for the three predictions


#### 3. Most Popular Source of Tweets.
During assessment, it was observed that there were various sources used to tweet by the WeRateDogs(@dogrates) account. From the Seaborn countplot, it is noted that an iPhone was most popular source of tweets for the account.



## Conclusion
The `archive_master` dataset was imported to answer three insight topics selected. The data was cleaned and the dataframe split into two sub-dataframes i.e `tweets_df` and `images_df` for easier analysis. This is because the `archive_master` dataset contains duplicated tweet_ids as a result of the multiple levels of predictions of the images.

For the first topic, descriptive statistics was carried out for the *retweet and favourite counts* columns in the `tweets_df` sub dataframe. A scatter plot was then plotted to visually confirm the correlation between the two retweets and favourites for the tweets in the WeRateDogs (@dogrates) account.

For the second topic, a bar plot was made to show the comparison in accuracy that the different levels of predictions yielded to successfully identify the breed of dogs in the images in the neural network.

For the third topic, a simple seaborn plot was made to show the source of tweets used frequently for the @dogrates account. The plot complemented the statistics obtained from the `value_counts()` function applied on the *source* column of the `tweets_df` sub dataframe.

The success of the EDA and Data visualization is owed to the wrangling phase of the data Analysis. All the issues that would have proved to be cumbersome had already been cleaned. Therefore, finding insights on the topics was very efficient.
